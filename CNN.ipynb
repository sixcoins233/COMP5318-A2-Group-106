```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import pickle
import math

# Checking for GPU availble
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
# pkl dataset loading
def load_pkl(file_path):
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    return data['data'], data['labels']
```

```python
def visualize_emnist_samples(data, labels):
    # get all the unique labels and number of it
    unique_labels = np.unique(labels)
    num_labels = len(unique_labels)
    
    # create a dict to store all the index for each unique label
    label_to_index = {}
    for label in unique_labels:
        label_to_index[label] = np.where(labels == label)[0][0]
        
    # calculate the suit size of diagram
    n_cols = math.ceil(math.sqrt(num_labels))
    n_rows = math.ceil(num_labels / n_cols)

    # graph style setting
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 15))
    fig.suptitle("EMNIST Samples (First image for each label)", fontsize=16)
    
    for i, label in enumerate(unique_labels):
        row = i // n_cols
        col = i % n_cols
        
        # get the first image for each label
        img = data[label_to_index[label]]

        # get current axes for eliminate the rest empty diagram
        if n_rows > 1:
            ax = axes[row, col]
        else:
            ax = axes[col]
        
        # plt
        axes[row, col].imshow(img, cmap='gray')
        axes[row, col].axis('off')
        axes[row, col].set_title(f"Label: {label}")

    # remove rest empty diagram
    for i in range(num_labels, n_rows * n_cols):
        row = i // n_cols
        col = i % n_cols
        if n_rows > 1:
            fig.delaxes(axes[row, col])
        else:
            fig.delaxes(axes[col])
    
    plt.tight_layout()
    plt.show()
```

```python
# load the dataset
x_train, y_train = load_pkl('emnist_train.pkl')
x_test, y_test = load_pkl('emnist_test.pkl')

# Visualize the original data
visualize_emnist_samples(x_train, y_train)
```

```python
# data preprocessing
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# check the shape of the dataset (samples, height, width, channels)
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

# model constract
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(62, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# model training
history = model.fit(x_train, y_train, epochs=10, 
                    validation_split=0.1,
                    batch_size=128)

# evaluate the model performance
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'\nTest accuracy: {test_acc}')

# plt the model training process
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# visualize the predict result
def visualize_predictions(model, x, y, num_images=25):
    predictions = model.predict(x[:num_images])
    fig, axes = plt.subplots(5, 5, figsize=(15, 15))
    for i, ax in enumerate(axes.flat):
        ax.imshow(x[i].reshape(28, 28), cmap='gray')
        true_label = y[i]
        pred_label = np.argmax(predictions[i])
        ax.set_title(f"True: {true_label}, Pred: {pred_label}")
        ax.axis('off')
    plt.tight_layout()
    plt.show()

visualize_predictions(model, x_test, y_test)
```

```python
# saving model if need (tensorflow formate)
model.save('emnist_model')
print("Model saved in TensorFlow format")
```
